```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# DATA PREPROCESSING

```python
dataset['PURPOSE'].fillna("NOT",inplace = True)
```

```python
dataset.head()
```

```python
dataset['START_DATE'] = pd.to_datetime(dataset['START_DATE'], errors = 'coerce')

dataset['END_DATE'] = pd.to_datetime(dataset['END_DATE'], errors = 'coerce')
```

```python
dataset.info()
```

```python
from datetime import datetime

dataset['DATE'] = pd.DatetimeIndex(dataset['START_DATE']).date
dataset['Time'] = pd.DatetimeIndex(dataset['START_DATE']).hour
```

```python
dataset.head()
```

```python
dataset['DAY_TIME'] = pd.cut(x=dataset['Time'],bins = [0,10,15,19,24],labels = ['Morning','Afternoon','Evening','Night'])
```

```python
dataset.head()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.shape
```

#DATA VISUALIZATION

```python
plt.figure(figsize=(20,5))

plt.subplot(1,2,1)

sns.countplot(dataset['CATEGORY'])

plt.subplot(1,2,2)
sns.countplot(dataset['PURPOSE'])
```

```python
sns.countplot(dataset['DAY_TIME'])
```

```python
dataset.head()
```

```python
dataset['MONTH'] = pd.DatetimeIndex(dataset['START_DATE']).month

month_label = {1.0: 'Jan', 2.0: 'Feb', 3.0: 'Mar', 4.0: 'April',
 5.0: 'May', 6.0: 'June', 7.0: 'July', 8.0: 'Aug',
 9.0: 'Sep', 10.0: 'Oct', 11.0: 'Nov', 12.0: 'Dec'}

dataset["MONTH"] = dataset.MONTH.map(month_label) # Number months ko string name
mon = dataset.MONTH.value_counts(sort=False) # Har month ke counts calculate ka
```

```python
dataset.head()
```

```python
df = pd.DataFrame({
 "MONTHS": mon.values, # Har month ka total count.
 "VALUE COUNT": dataset.groupby('MONTH', sort=False)['MILES'].max() # Har mo
})
p = sns.lineplot(data=df) # Line plot banata hai.
p.set(xlabel="MONTHS", ylabel="VALUE COUNT") # Axis labels set karta ha
```

```python
dataset['DAY'] = dataset.START_DATE.dt.weekday
day_label = {
 0: 'Mon', 1:'Tues', 2:'Wed', 3:'Thur',4:'Fri', 5:'Sat', 6:'Sun'}
dataset['DAY'] = dataset['DAY'].map(day_label)
```

```python
dataset.head()
```

```python
day_label = dataset.DAY.value_counts()

sns.barplot(x=day_label.index, y= day_label)
plt.xlabel('DAY')
plt.ylabel('COUNT')
```
